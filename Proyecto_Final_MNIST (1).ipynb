A continuación, se presenta el EDA realizado tomando como guía las preguntas en la sección de 'Metodología'.

> El arreglo `y` contiene las etiquetas de las imágenes en `x`. **Cada etiqueta indica el número que representa la imagen correspondiente**, por ejemplo, si `y[0]` es igual a 5, entonces la primera imagen en `x` representa el número 5.

5. **Muestre como imagen una instancia de cada clase**, de forma que se valide que existen diferencias entre estas.

````python
# Diccionario para almacenar una instancia de cada clase
instancias = {}

# Iterarción sobre el conjunto de entrenamiento
for i in range(len(x_train)):
    # si no hemos visto una instancia de la clase correspondiente
    if y_train[i] not in instancias:
        # almacenar la imagen correspondiente
        instancias[y_train[i]] = x_train[i]
    
    # si hemos visto una instancia de cada clase, detener el ciclo
    if len(instancias) == 10:
        break

# Mostrar una instancia de cada clase
fig, ax = plt.subplots(2, 5, figsize=(10, 5))
for i in range(10):
    row = i // 5
    col = i % 5
    ax[row, col].imshow(instancias[i], cmap='gray')
    ax[row, col].set_title(f'Clase {i}')
    ax[row, col].axis('off') # escondiendo los ticks del axis
plt.show()
````

## Preprocesamiento

````python
# Escalamiento para llevarlo de 0-255 a 0-1

# Se encontró el máximo de cada arreglo
max_value_train = x_train.max()
max_value_test = x_test.max()

# Se escaló en base a ese máximo
x_train_scaled = x_train / max_value_train
x_test_scaled = x_test / max_value_test
````

## Clasificador base

````python
import tensorflow as tf

# Importando el módulo de keras
from tensorflow import keras
````

````python
# Definición del modelo
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='sigmoid'),
    keras.layers.Dense(10, activation='softmax')
])
````

````python
# Compilación del modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
````

````python
# Entrenamiento del modelo
model.fit(x_train_scaled, y_train, epochs=10, batch_size=64, verbose=0)
````

````python
# Evaluación del modelo
test_loss, test_acc = model.evaluate(x_test_scaled, y_test, verbose=0)
print('Test accuracy:', test_acc)
````

````python
# Predicciones
classifications = model.predict(x_test_scaled)
print('Primera entrada en las clasificaciones:', classifications[0])
````

> La primera entrada en las clasificaciones es un vector de 10 números, que representan la probabilidad de que la imagen de prueba pertenezca a cada una de las 10 clases posibles (números del 0 al 9).

> La suma de todas las entradas del vector es 1. Esto se debe a que la capa de salida utiliza una función de activación **softmax**, que garantiza que la salida de la red sea una distribución de probabilidad.

> En otras palabras, la red asigna una probabilidad a cada una de las clases posibles, y la clase con la probabilidad más alta se considera la predicción de la red.

> La clase correspondiente a la primera entrada se puede obtener utilizando la función argmax de NumPy:

````python
first_class = np.argmax(classifications[0])
print('Clase de la primera entrada:', first_class)
````

````python
# Visualización de la primera imagen del conjunto de prueba
plt.imshow(x_test[0], cmap='gray')
plt.axis('off')
plt.show()
````

Cuando se aumentan los epochs en un modelo de red neuronal aveces el valor de pérdida deja de disminuir y aumenta, esto se debe a un sobreajuste o "overfitting" del modelo a los datos de entrenamiento. Como el modelo tiene más oportunidades de ajustarse a los datos de entrenamiendo se sobreajusta, como se puede ver en el gráfico de 30 epochs. ###Early stopping

````python
# Definición del modelo
model_exp7 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='sigmoid'),
    keras.layers.Dense(10, activation='softmax')
])

# Compilación del modelo
model_exp7.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Early Stopping Callback
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=5)

# Entrenamiento del modelo
history_exp7 = model_exp7.fit(x_train_scaled, y_train, epochs=50, batch_size=64,
                              verbose=0, validation_split=0.2, 
                              callbacks=[early_stop])
````

````python
history_data_exp7 = pd.DataFrame(history_exp7.history)
history_data_exp7.loc[:, ['loss', 'val_loss']].plot();
plt.title('Curvas de entrenamiento y validación')
````

````python
# Evaluar el modelo en el conjunto de prueba
test_loss_exp7, test_acc_exp7= model_exp7.evaluate(x_test_scaled, y_test,verbose=0)

# Imprimir la precisión y la pérdida en el conjunto de prueba
print('Precisión en el conjunto de prueba:', test_acc_exp7)
print('Pérdida en el conjunto de prueba:', test_loss_exp7)
````

Con la técnica Early Stopping se detuvo el entrenamiento antes de que se sobreajustara el modelo. Esto ayuda a mejorar la capacidad de generalización de la red neuronal evitando que se ajuste a los datos de entrenamiento. Con ayuda de la gráfica se visualiaza como las curvas se están uniendo y bajando lo que significa que se arregló el problema de sobreajuste que tenía la gráfica anterior.

**Escriba un breve análisis comparando la utilización de unidades sigmoid y ReLU.** Explicar por qué pueden darse estas diferencias. 

Tanto las unidades sigmoid como las unidades ReLU son funciones de activación comunes utilizadas en las redes neuronales, pero son utilizadas para diferentes propósitos. La función sigmoide generalmente se usa en casos donde la variable de salida es de tipo binaria, lo que signifca que puede generar problemas de saturación cuando se les da una entrada muy grande. La función ReLU no tiene este problema, esta devuelve valores positivos para entradas positivas y 0 para entradas negativas lo que puede llevar a una mayor eficiencia en el entrenamiento de la red neuronal.

La función de activación ReLu según estudios ha demostrado ser más efectiva que la función de activación sigmoide, debido a su simplicidad y que aceleran la convergencia del entrenamiento. En este caso, la precisión con la función ReLu fue menor que con sigmoide y las pérdidas son menores con la función sigmoide que con Relu. En términos de los gráficos se puede ver como el experimento con la función ReLu muestra que el modelo se encuentra sobreajustado y con la función sigmoide no ocurre esto. En conclusión nos conviene usar la función sigmoide.
